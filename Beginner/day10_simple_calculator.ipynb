## Project - Simple Calculator

This is a simple calculator with basic functions to `add, substract, multply and divide`.
Below code also covers implemntation of `for loop`, `while loop`, `if else statements`, defining a `function` and concept of `recursion`

```python
# Calculator
#import art
# added logo for calculator to improve user experiecnce when they are using it it always start with welcoming them and logo
logo = """
 _____________________
|  _________________  |
| | Pythonista   0. | |  .----------------.  .----------------.  .----------------.  .----------------. 
| |_________________| | | .--------------. || .--------------. || .--------------. || .--------------. |
|  ___ ___ ___   ___  | | |     ______   | || |      __      | || |   _____      | || |     ______   | |
| | 7 | 8 | 9 | | + | | | |   .' ___  |  | || |     /  \     | || |  |_   _|     | || |   .' ___  |  | |
| |___|___|___| |___| | | |  / .'   \_|  | || |    / /\ \    | || |    | |       | || |  / .'   \_|  | |
| | 4 | 5 | 6 | | - | | | |  | |         | || |   / ____ \   | || |    | |   _   | || |  | |         | |
| |___|___|___| |___| | | |  \ `.___.'\  | || | _/ /    \ \_ | || |   _| |__/ |  | || |  \ `.___.'\  | |
| | 1 | 2 | 3 | | x | | | |   `._____.'  | || ||____|  |____|| || |  |________|  | || |   `._____.'  | |
| |___|___|___| |___| | | |              | || |              | || |              | || |              | |
| | . | 0 | = | | / | | | '--------------' || '--------------' || '--------------' || '--------------' |
| |___|___|___| |___| |  '----------------'  '----------------'  '----------------'  '----------------' 
|_____________________|
"""
print(logo)
#Add - created function for addition
def add(n1,n2):
  return n1+n2

#Subtract - created function for substraction
def sub(n1,n2):
  return n1-n2

# Multiply - created function for multiplication
def mul(n1,n2):
  return n1 * n2

# Divide - created function for division
def div(n1,n2):
  return n1 / n2

# created dictionary named operation with key as symbol and value as the function that we created earlier
operations = {
  "+": add,
  "-": sub,
  "*": mul,
  "/": div
}



def calculation():
  #Asks user for input of initial number
  num1 = float(input("What's the number?: "))
  # We use for loop on dictionary and print out every element to key to let user know what kind of operation can be performed
  for x in operations:
    print(x)
  #Created a simple variable called end_calc to later use ot to initiate and end while loop
  end_calc = True

  #while loop initiated for all below logic
  while end_calc:
    #we ask user to pick a operation they want to perform 
    operation_symbol = input("Pick an operation: ")
    #we ask user to input another number to perform the selected operation
    num2 = float(input("What's the next number?: "))
    # we create a variable answer to store the calculated answer
    # we tap in operations dictionary and look for the key which operation symbol is the key that we took as an input from user
    # This will trigger out function that we created as the value assigned to keys are the function and in () we pass the input for function
    answer = operations[operation_symbol](num1,num2)
    
    # we print the required output for calculation
    print(f"{num1} {operation_symbol} {num2} = {answer}")
    
    # created varibale named should_continue which take input from user
    # user get a choice if they want to continue withy the above answer that we have already calculated
    # if not if they want to start a whole new fresh calculation or end the calculator
    should_continue = input(f"Do you want to continue calculation with {answer} then type 'yes' or type 'no' if you want to start a fresh new calculation or type 'end' to close calculator: ").lower()
    
    #if user inputs yes we assign answer to num1
    if should_continue == 'yes':
      #assigning answer to variable 1 and again we tap in from start we num1 as answer for further calculation
      num1 = answer
    #if user types no
    elif should_continue == 'no':
      #we assign False to varibale end_calc which will eventually end the while loop 
      end_calc = False
      #then we again recall calculation function which will start a fresh new calculation
      
      #this will work as recurion as we are callinf fucntion within a funcction
      calculation()
    #if user selects end then it will go in this else statement
    else:
      # we assign false to end_calc which will end the while loop and we come out loop
      end_calc = False
      # we print Goodbye just before we close loop and come out of it
      print("Goodbye!")
#as we created this funbction and recalled this function within function in earlier stage which works as recursion to tap in all logic we need to run this once to strat the program initially
calculation()
```
