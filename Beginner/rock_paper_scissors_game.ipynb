```python
import random

print('''
Welcome to the ultimate challenge of rock, paper and scissors.
Can you beat the computer!
Have a try then. :)
''')

users_choice = int(input('''
What do you choose? Rock(0), Paper(1) or Scissors(2). 
Choose a number 0 , 1 or 2
'''))


rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
 '''

scissors = '''
     _______
 ---'   ____)____
          ______)
        __________)
       (____)
---.__(___)
 '''

if users_choice >=3 or users_choice < 0:
    print("You typed an invalid number, you lose!")

    
else:
    
    
    hand_gesture = [rock,paper,scissors]
    players_choice = hand_gesture[users_choice]
    computers_choice = random.randint(0,2)
    computer_choose = hand_gesture[computers_choice]
    
    
    print(f'''
    You Choose: {players_choice}
    Computer Choose: {computer_choose}
    ''')
    
    
    if users_choice == 0 and computers_choice == 2:
        print('You Win!')
    elif computers_choice == 0 and users_choice == 2:
        print("You Lose")
    elif computers_choice > users_choice:
        print("You Lose")
    elif users_choice > computers_choice:
        print("You Win!")
    elif users_choice == computers_choice:
        print("It's a Draw")
```
