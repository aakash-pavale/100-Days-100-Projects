## 3.1 Leap Year
- The goal here is to calcualte whether the Year inputted by the user is a leap year. 
- This is how you work out whether if a particular year is a leap year.

    on every year that is evenly divisible by 4 

    **except** every year that is evenly divisible by 100 

    **unless** the year is also evenly divisible by 400

    e.g. The year 2000:

    2000 รท 4 = 500 (Leap)

    2000 รท 100 = 20 (Not Leap)

    2000 รท 400 = 5 (Leap!)

    So the year 2000 is a leap year.

### Use the following flow chart to write the if else statement:

```python
# taking input from user
year = int(input("Which year do you want to check?\n"))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print('\nThis is a Leap year.')
        else:
            print('\nThis is not a Leap year.')
    else:
        print('\nThis is a Leap year.')
else:
    print('\nThis is not a Leap year.')
```

- To check whether a number is cleanly divisible, we use `%` symbol called as modulus, which calculates the remainder of the division operation that we are performing.

## 3.2 Pizza Bill Generator

Here, we will be evaluating the bill of the pizza based on the size a user selects and the add-ons he/she prefers to take.

- <p>The cost of Small, Medium and Large pizzas are \$15, \$20 and \$25 respectively



- <p>If the user prefers to add pepperoni, then the add-on cost would be \$2, if its a Small Pizza and \$3 if its a Medium or Large Pizza. 


- Extra Cheese would cost $1 regardless the size of pizza

```python
# taking inputs from user
print("Welcome to Python Pizza Deliveries!")
size = input("\nWhat size pizza do you want? S, M, or L: ")
add_pepperoni = input("\nDo you want pepperoni? Y or N: ")
extra_cheese = input("\nDo you want extra cheese? Y or N: ")
```

Declaring a counter variable which we will be using to add the bill amount based on different conditions, which will then eventually give us the final bill amount as our logic flows below

```python
bill = 0
```

Adding the Pizza's amount to our counter based on user selection:

```python
if size == 'S':
    bill += 15
elif size == 'M':
    bill += 20
else:
    bill += 25
    
# The 'bill+=15' statement adds 15 to the oroginally declared variable where bill was 0
# We don't need to add condition for Large Pizza as the 'else' statement takes care of it
```

Adding condition for the add-on of pepperoni:

```python
if add_pepperoni == 'Y':
    if size == 'S':
        bill += 2
    else:
        bill += 3
        
# Here, we don't need to add conditions where the user doesn't want to add pepperoni as 
# the amount for this add-on will only get added to the bill if the condition is true.
# Also, we have added a nested condition if the user wants pepperoni and the pizza size is Small, 
# the 'else' statement takes care of Medium and Large Pizza as they both have the same add-on amount for pepperoni
```

Adding condition for the add-on of extra cheese:

```python
if extra_cheese == 'Y':
    bill += 1

print(f"Your final bill based on your preferences is: ${bill}")
```

## 3.3 Roller Coaster Ticket

This program helps us generate the ticket price based on user's height, age and whether he wants a picture taken.
- If the the user height in cm is less than 120, then he should not get a ticket and the further questions should not be asked like what the age is and whether he/she wants a photo taken.

- Ticket Prices are as follows:
    - Age < 12, then \$5
    - Age between 12 and 18, then \$7
    - Age > 18, then \$12

- If he/she wants a photo taken then additional \$3 should be added to the bill

```python
# taking height input
height = int(input('What is your height in cm?\n'))

# making control flow based on logic discussed above
if height >= 120:
    print('You can ride the roller coaster!')           # letting user know he/she can ride
    age = int(input("\nWhat is your age?\n"))           # asking for age only if the height > 120cm
    if age < 12:
        bill = 5                                        # declaring bill cost based on age criteria
        print("Child tickets cost: $5.")                # displaying ticket cost
    elif age >=12 and age <=18:
        bill = 7
        print("Teen tickets cost: $7")
    else:
        bill = 12
        print("Adult tickets cost: $12")

    photo = input("\nDo you want a picture? Y or N\n")    # asking whether they want a photo
    if photo == 'Y':
        bill += 3                                       # adding $3 to bill amount based on selection made
        print(f"\nYour ticket cost is ${bill}.")
    else:
        print(f"\nYour ticket cost is ${bill}.")

else:
    print('\nYou cannot ride the roller coaster!, Mummy ko bolo Complan pilaye!') #if the height criteria doesn't work
    
    
# The identation matters a lot to understand the flow of logic and to understand 
# where to put in the next condition and the execution of the logic
```
