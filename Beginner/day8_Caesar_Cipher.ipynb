Well its works on a simple principle:

`In cryptography, a Caesar cipher, also known as Caesar's cipher, the shift cipher, Caesar's code or Caesar shift, is one of the simplest and most widely known encryption techniques. It is a type of substitution cipher in which each letter in the plaintext is replaced by a letter some fixed number of positions down the alphabet. For example, with a left shift of 3, D would be replaced by A, E would become B, and so on. The method is named after Julius Caesar, who used it in his private correspondence.
The encryption step performed by a Caesar cipher is often incorporated as part of more complex schemes, such as the Vigenère cipher, and still has modern application in the ROT13 system. As with all single-alphabet substitution ciphers, the Caesar cipher is easily broken and in modern practice offers essentially no communications security.`

```python
# Step 2 defining a function to take the parameters as an input which later we will be passing argument in
def caesar(start_text, shift_amount, cipher_direction):
  end_text = "" # empty string

# we create a if statement just to make shift negative whenever uses passes argument in parameter ass decode
  if cipher_direction == "decode": 
    shift_amount *= -1
    
# here we create a for loop to check each char in start_text and check that each char in aplhabet list
  for char in start_text:
    if char in alphabet:
      position = alphabet.index(char) # capturing its index in variable
      new_position = position + shift_amount # adding shift to the possition to either encode or decode
      end_text += alphabet[new_position] # Then we just concate every string together for an output in end_text
    
    else: # we use else here in case there are inputs which are not in our alphabet list we just simply concat them to our above result
      end_text += char
  print(f"Here's the {cipher_direction}d result: {end_text}")
 
# Step 4 Print logo
print(logo)
```

```python
# Step 4
should_end = False # creating a variable with boolean value False to use it in loop

while not should_end: # while loop will continue unless the should_end becomes True

# we create input parameters for user to so that they can pass argument in it.
  direction = input("Type 'encode' to encrypt, type 'decode' to decrypt:\n")
  text = input("Type your message:\n").lower()
  shift = int(input("Type the shift number:\n"))
# adding a precaution layer in case user puts a large shift value which is greater than 26
  if shift > 26:
        shift = shift % 26
# Calling the caesar function to give ouput as per the users input to parameters.
  caesar(start_text=text, shift_amount=shift, cipher_direction=direction)

# created a variable named restart and asking user for input yes or no in case they want to continue encoding/decoding
  restart = input("Type 'yes' if you want to go again. Otherwise type 'no'.\n")
  if restart == "no": # as in this copndition user inputs no varibale should_end changes to true while loop ends.
    should_end = True
    print("Goodbye") # program will say good and end the loop.
```
